```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

def car_price_regression(filename):
    df = pd.read_csv(filename)

    features = ['Year', 'Engine HP', 'Engine Cylinders', 'highway MPG', 'city mpg', 'Popularity']
    target = 'MSRP'
    X = df[features]
    y = df[target]

    X = X.fillna(X.mean())

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

    model = LinearRegression()
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    rmse_lr = mean_squared_error(y_test, y_pred, squared=False)

    y_dummy = np.full_like(y_test, y_train.mean(), dtype=np.float64)

    rmse_dummy = mean_squared_error(y_test, y_dummy, squared=False)

    print(f"RMSE линейной регрессии: {rmse_lr:.2f}")
    print(f"RMSE глупой модели: {rmse_dummy:.2f}")
    print(f"Разница: {rmse_dummy - rmse_lr:.2f}")

car_price_regression('data.csv')
```
