```python
import pandas as pd
from scipy.stats import ttest_ind

def website_analyse(file_path):
    df = pd.read_csv(file_path, delimiter=',')

    df['Просмотренные страницы'] = pd.to_numeric(df['Просмотренные страницы'], errors='coerce')
    df['Время на сайте (сек)'] = pd.to_numeric(df['Время на сайте (сек)'], errors='coerce')

    mean_pages = df['Просмотренные страницы'].mean()

    std_time = df['Время на сайте (сек)'].std()

    mean_time = df['Время на сайте (сек)'].mean()
    coeff_var = std_time / mean_time if mean_time != 0 else 0
    if coeff_var > 0.25:
        variability = 'Высокая вариабельность'
    elif coeff_var < 0.1:
        variability = 'Слабая вариабельность'
    else:
        variability = 'Умеренная вариабельность'

    mobile_pages = df[df['Тип устройства'] == 'Мобильное']['Просмотренные страницы']
    desktop_pages = df[df['Тип устройства'] == 'Десктоп']['Просмотренные страницы']
    t_stat, p_value = ttest_ind(mobile_pages, desktop_pages, equal_var=False)

    if p_value > 0.05:
        stat_diff = 'Нет статистически значимой разницы'
    else:
        stat_diff = 'Есть статистически значимая разница в количестве страниц'

    print(f"Среднее количество просмотренных страниц таблиц: {mean_pages:.2f}")
    print(f"Стандартное отклонение: {std_time:.2f}")
    print(f"{variability}: {coeff_var:.2f}")
    print(f"{stat_diff}: {p_value:.4f}")

website_analyse('website_visits_december_2024.csv')
```
