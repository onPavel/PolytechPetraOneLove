```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('books.csv')

def plot_books_2023(ax, df):
    df_2023 = df[df['Год'] == 2023]
    ax.plot(df_2023['Месяц'], df_2023['Книги'], marker='o', color='red', label='Прочитано книг')
    ax.set_title('Прочитанные книги 2023')
    ax.set_xlabel('Месяцы')
    ax.set_ylabel('Количество книг')
    ax.grid(True)
    ax.legend()

def plot_pages_2024(ax, df):
    df_2024 = df[df['Год'] == 2024]
    ax.bar(df_2024['Месяц'], df_2024['Страницы'], color='green', label='Прочитано страниц')
    ax.set_title('Прочитанные страницы 2024')
    ax.set_xlabel('Месяцы')
    ax.set_ylabel('Количество страниц')
    ax.grid(True, axis='y')
    ax.legend()

fig, axs = plt.subplots(1, 2, figsize=(14, 6))

plot_books_2023(axs[0], df)
plot_pages_2024(axs[1], df)

plt.tight_layout()
plt.show()
```
